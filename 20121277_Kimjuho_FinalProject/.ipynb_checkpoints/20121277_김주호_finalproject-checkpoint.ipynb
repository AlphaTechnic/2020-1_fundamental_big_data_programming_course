# cafehero_sales.txt 읽어오기

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import re 
import datetime as dt

# cafehero_sales.txt 에 있는 데이터를 가져와 DataFrame을 생성하기 쉬운 형태로 data_list에 담는다.
data_list = []    

# 해당 파일을 읽어오는 것을 실패했을 경우에 대한 예외처리를 해준다.
try:
    with open("C:\\Users\\juho3\\Desktop\\Python\\Fundamental_BigData_programming_course\\final_project\\cafehero_sales.txt",
              'rt', encoding = 'utf-8') as fp:
        
        # cafehero_sales.txt 파일의 첫번째 line이 DataFrame의 column이 될 것이다.
        column_name = fp.readline().split()
           
        while True:
            line = fp.readline()
            if line == '':
                break
                
            # 텍스트파일 속 원본 데이터가 1,000,000 처럼 쉼표가 있어서, 이를 제거하는 작업을 정규표현식을 이용해 해주며,
            # refined된 data를 리스트에 담는다.
            line = re.sub(r"[,]", "", line)
            data = line.split()
                        
            # 각 line별 첫번 째 단어인 '2016-09' 같은 날짜는 DataFrame의 index의 역할을 할 것이므로, 문자열 형태로 그대로 담는다.
            # 그 뒤에 오는 n번 째 단어들은 int의 형태로 바꿔 line_refined 리스트에 담고, 
            # 이 완성된 line_refined 리스트를 data_list에 담는다.
            line_refined = []
            for ind, content in enumerate(data):
                if ind == 0:
                    line_refined.append(content)
                else:
                    line_refined.append(int(content))

            data_list.append(line_refined)
            
        print("해당 파일의 Data를 정상적으로 가져왔습니다.")
        
except FileNotFoundError:
    print("파일 읽기 Error!!")
    print("해당 파일을 찾을 수 없습니다.")
    
```

```python
# data_list의 원소로 있는 line_refined 리스트의 첫 번째원소(=날짜)가 DataFrame의 index가 된다.
ind = [data_list[i][0] for i in range(len(data_list))]
cafehero = pd.DataFrame(data = data_list, columns = column_name, index = ind)
# index로 설정되어 있는 '날짜'가 첫번째 첫번째 column으로 중복되어 입력되어 있으므로 이를 제거
del cafehero['영업월']

# 새로운 지표인 "순매출", "객단가", "현금매출 비율"을 생성.
# 객단가 <- 한 팀이 온다면, 돈을 얼마나 쓰고 가는지에 대한 지표
cafehero['순매출'] = cafehero['총매출'] - (cafehero['할인'] + cafehero['부가세'])
cafehero['객단가'] = list(map(int, cafehero['총매출'] / cafehero['고객수']))
cafehero['현금매출비율'] = round(cafehero['현금영수금액'] / (cafehero['현금영수금액'] + cafehero['카드영수금액']) * 100, 2)

# 쓰지 않을 지표들은 삭제
del cafehero['취소']
del cafehero['취소금액']
del cafehero['현금취소건수']
del cafehero['현금취소금액']
del cafehero['카드취소건수']
del cafehero['카드취소금액']

# print는 행과 열을 맞춰 보여주지 못하여 head 를 사용하여 DataFrame을 보인다.
cafehero.head(len(cafehero))
```

```python
print("***매장의 규모를 먼저 점검합니다***")

# 29개월 간 순매출, 현금매출, 카드매출의 평균을 구한다.
aver_sales = int(sum(cafehero['총매출']) / len(cafehero['총매출']))
aver_sales_by_cash = int(sum(cafehero['현금영수금액']) / len(cafehero['현금영수금액']))
aver_sales_by_card= int(sum(cafehero['카드영수금액']) / len(cafehero['카드영수금액']))

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 평균액수를 bar그래프로 표현
index = ['Total', 'Cash', 'Card']
vals = [aver_sales, aver_sales_by_cash, aver_sales_by_card]
ax1.bar(index, vals)

# 카드매출과 현금매출의 비율을 원그래프로 표현
ScaleOfStore = [aver_sales_by_cash, aver_sales_by_card]
label = ['Cash', 'Card']
ax2.pie(ScaleOfStore, labels = label, explode = (0, 0.1), autopct = '%1.1f%%', shadow = True)

plt.suptitle("Average Monthly NET Sales for 29 Months", y = 1 , size = 14) 
plt.show()

print(">> 매장의 %s개월 간 평균 순매출액은 약 %d원 입니다.\n\n\n\n" %(len(cafehero['총매출']), aver_sales))

############################################################################################

# 순매출 column의 가장 최근(2018-10) 기록에 접근하여, 위와 같은 형태의 그래프를 생성한다.
latest_sales = cafehero['총매출'].loc['2018-10']
latest_sales_by_cash = cafehero['현금영수금액'].loc['2018-10']
latest_sales_by_card= cafehero['카드영수금액'].loc['2018-10']

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

index = ['Total', 'Cash', 'Card']
vals = [latest_sales, latest_sales_by_cash, latest_sales_by_card]
ax1.bar(index, vals)

ScaleOfStore = [latest_sales_by_cash, latest_sales_by_card]
label = ['Cash', 'Card']
ax2.pie(ScaleOfStore, labels = label, explode = (0, 0.1), autopct = '%1.1f%%', shadow = True)

plt.suptitle("Latest Average Monthly NET Sales", y =1, size = 14) 
plt.show()

print(">> 매장의 가장 최근 한달 평균 총매출액은 약 %d원 입니다." %(latest_sales))
```

```python
# 해당 column data들을 0~1로 scaling하는 함수를 정의한다.
# 이 scaling을 이용하여 y값을 조정하면, 서로 다른 지표들의 추이에 있어 유사성이 있는지 비교할 수 있게 된다.

def scaling(column):
    maximum = max(column)
    minimum = min(column)
    column_scaled0to1 = (column - minimum) / (maximum - minimum)
    
    return column_scaled0to1
```

```python
print("***29개월 간 '방문 팀 수'와 '순매출'의 추이를 살핍니다***")

fig = plt.figure(figsize = (25,6))
plt.title('Number of Teams visited')
plt.plot(cafehero['고객수'], 'bo-')
plt.xticks(rotation = 30)
plt.show()

fig = plt.figure(figsize = (25,6))
plt.title('Net Sales')
plt.plot(cafehero['순매출'], 'ro-')
plt.xticks(rotation = 30)
plt.ylabel('won', fontsize = 15)
plt.show()


# y값을 0~1로 스케일링하여 두 그래프를 합쳐 보인다.
fig = plt.figure(figsize = (25,6))
plt.plot(scaling(cafehero['고객수']), 'bo-')
plt.plot(scaling(cafehero['순매출']), 'ro-')
plt.legend(['Number of Teams visited', 'Net Sales'])
plt.xticks(rotation = 30)
plt.show()

print(">> 매장의 '방문 팀 수'와 '순매출'이 꾸준한 증가 추세임을 알 수 있습니다.")
```

```python
print("***29개월 간 '카드매출'과 '현금매출'의 추이를 살핍니다***")

fig = plt.figure(figsize = (25,6))
plt.title('Number of Receipts by Card')
plt.plot(cafehero['카드영수건수'], 'ko-')
plt.xticks(rotation = 30)
plt.show()

fig = plt.figure(figsize = (25,6))
plt.title('Number of Receipts by Cash')
plt.plot(cafehero['현금영수건수'], 'go-')
plt.xticks(rotation = 30)
plt.show()

print(">> 매장의 '카드영수건수'와 '현금영수건수'가 꾸준한 증가 추세임을 알 수 있습니다.\n\n\n\n")

#######################################################################################################################
fig = plt.figure(figsize = (25,6))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

ax1.plot(cafehero['카드영수건수'] , 'ko-')
ax1.plot(cafehero['현금영수건수'] , 'go-')
ax2.plot(cafehero['현금매출비율'], 'co-', alpha = 0.3)
ax1.legend(['Number of Receipts by Card', 'Number of Receipts by Cash'])
ax2.legend(['Share of Cash Payment'], loc = 'right')
ax2.set_ylabel('%', color = 'c', fontsize = 15)
ax2.tick_params('y', colors='c')

ax1.tick_params('x', rotation = 30)
ax1.annotate('Here!', color = 'r', xy = ('2016-06', 105), xytext = (0.02, 0.3), textcoords = 'axes fraction', fontsize = 20,\
             arrowprops =  dict(arrowstyle = '->', color = 'r'))

plt.show()

print(">> 매장의 영업 첫 달에는 '현금영수건수'(%s건)와 '카드영수건수'(%s건)가 거의 같았습니다." 
      %(cafehero['현금영수건수'].loc['2016-06'], cafehero['카드영수건수'].loc['2016-06']))
print(">> 이 후에 '현금영수건수'의 증가폭에 비하여 '카드영수건수'의 증가폭이 훨씬 큽니다.")
print(">> '현금매출비율'(Share of Cash Payment)이 큰 폭으로 감소합니다.\n")

#######################################################################################################################
# y값을 0~1로 스케일링하여 여러 그래프를 합쳐 보인다.
fig = plt.figure(figsize = (25,3))
plt.plot(scaling(cafehero['총매출']), 'ro-')
plt.plot(scaling(cafehero['카드영수건수']), 'ko-')
plt.plot(scaling(cafehero['현금영수건수']), 'go-')

plt.annotate('Here!', color = 'r', xy = ('2016-08', 0.65), xytext = (0.2, 0.8), textcoords = 'axes fraction', fontsize = 20,\
             arrowprops =  dict(arrowstyle = '->', color = 'r', alpha = 0.5, lw = '5'))  
plt.annotate('Here!', color = 'r', xy = ('2018-06', 0.45), xytext = (0.72, 0.2), textcoords = 'axes fraction', fontsize = 20,\
             arrowprops =  dict(arrowstyle = '->', color = 'r', alpha = 0.5, lw = '5'))  
plt.legend(['Number of Teams visited', 'Number of Receipts by Card', 'Number of Receipts by Cash'])
plt.xticks(rotation = 30)
plt.show()
print(">> '카드영수건수'는 후반으로 갈 수록 '총매출'과의 이격도가 거의 없어집니다.")
print(">> '현금영수건수'는 후반으로 갈 수록 '총매출'과의 이격도가 커집니다.\n")

print(">> 이는 현금매출이 매장에 기여하는 비율이 점점 낮아지고 있음을 직관적으로 보여줍니다.")
print(">> 좀 더 적극적으로 해석한다면, 카드를 많이 쓰는 연령대(20~40대)의 유입으로 매장이 성장했다고 분석할 수 있습니다.")
```

```python
from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests

with open('averages_monthly_temperature.txt', 'wt') as fp:
    for year in range(2016, 2019):
        # 입력한 url의 웹페이지의 소스코드를 가져온다.
        html = requests.get("https://www.weather.go.kr/weather/climate/past_table.jsp?stn=108&yy=" + str(year) + "&obs=07&x=29&y=5")

        # 소스코드를 tree구조로 parsing한다.
        soup = bs(html.text, 'html.parser')

        # findAll 함수를 사용하여, 'tr' 태그명의 data에 접근한다.
        data1 = soup.findAll('tr')
        
        # 정규표현식을 이용하여, 원하는 정보를 추출한다. 원하는 정보는 data[-1]에 있다.
        pattern = re.compile("(?<=<td>).+(?=<\/td>)")
        temperatures = pattern.findall(str(data1[-1]))

        # 2016년은 6월달부터의 정보를 text 파일에 입력,
        # 2017년은 모든 달의 정보를 text 파일에 입력,
        # 2018년은 10월달까지의 정보를 text파일에 입력
        if year == 2016:
            fp.write('\n'.join(temperatures[5:]))
            fp.write("\n")
        elif year == 2017:
            fp.write('\n'.join(temperatures))
            fp.write("\n")
        else:
            fp.write('\n'.join(temperatures[:10]))       

```
