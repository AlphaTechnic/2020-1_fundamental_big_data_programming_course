# pie chart 그리기 실습

```python
import matplotlib.pyplot as plt

data = [26.2, 26.9, 16.2, 30.6]
languages = ['java', 'Python', 'C++', 'Ruby']

# 아래 plot시 처리한 내용 : (순서대로) slice labels, auto-labeling the percentage, offsetting a slice, drop-shadow, custom start angle 
plt.pie(data, labels = languages, autopct = '%1.1f%%', explode = (0, 0.1, 0, 0), shadow = True, startangle = 45)
plt.show()
```

# 이미지만 출력하기 실습

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("민들레꽃.jpg")
plt.imshow(img)
plt.axis('off')  # 축의 값을 숨긴다.
plt.show()
```
