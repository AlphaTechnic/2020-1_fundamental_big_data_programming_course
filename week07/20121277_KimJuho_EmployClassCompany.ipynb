```python
%%writefile Employee_class.py

class Person:
    def __init__(self, name, age, gender):
        self._name = name
        self._age = age
        self._gender = gender
    
    def aboutMe(self):
        print("이름 :", self._name, ", 나이 :", self._age, ", 성별 :", self._gender)
        
    @property   
    def name(self):
        return self._name
    
    def get_age(self):
        return self._age
    def set_age(self, age):
        self._age = age
    age = property(get_age, set_age)
        
class Employee(Person):
    def __init__(self, name, age, gender, hiredate, teamname, phone_num):
        Person.__init__(self, name, age, gender)
        self._hiredate = hiredate
        self._teamname = teamname
        self._phone_num = phone_num
    
    def timechecker(func):
        import time
        def checktime(self):
            cur_time = time.strftime("%Y-%m-%d %H:%M" ,time.localtime())
            print(cur_time, end = ' ')
            func(self)
        return checktime
    
    @timechecker        
    def startWork(self):
        print("출근!")
        print("출근하였습니다.\n")    
    @timechecker
    def endWork(self):
        print("퇴근!")
        print("퇴근합니다.\n")
        
    def get_hiredate(self):
        return self._hiredate
    def set_hiredate(self, date):
        self._hiredate = date
    hiredate = property(get_hiredate, set_hiredate)
    
    def get_teamname(self):
        return self._hiredate
    def set_teamname(self, name):
        self._hiredate = name
    teamname = property(get_teamname, set_teamname)
    
    def get_phone_num(self):
        return self._phone_num
    def set_phone_num(self, num):
        self._phone_num = num
    phone_num = property(get_phone_num, set_phone_num)
    
    def aboutMe(self):
        Person.aboutMe(self)
        print("고용일 :", self._hiredate, ", 부서 :", self._teamname, ", 휴대폰 번호 :", self._phone_num)
    
class Manager(Employee):
    manager_num = 0
    
    def __init__(self, name, age, gender, hiredate, teamname, phone_num, salary, manager_conference_day):
        Employee.__init__(self, name, age, gender, hiredate, teamname, phone_num)
        self._salary = salary
        self._manager_conference_day = manager_conference_day
        Manager.manager_num += 1
        print(self._name, "님이", self._teamname, "팀에 입사하였습니다.", "\n총", Manager.manager_num, "명의 팀장이 근무하게 되었습니다.\n")
        
    def increase_salary(self, val):
        self._salary += val
        print(val, "만원", "월급 인상!^^\n")
        
    def decrease_salary(self, val):
        self._salary -= val
        print(val, "만원", "월급 삭감!ㅜ\n")
        
    def get_salary(self):
        return self._salary
    def set_salary(self, val):
        self._salary = val
    salary = property(get_salary, set_salary)
    
    def aboutMe(self):
        Employee.aboutMe(self)
        print("월급 :", self._salary, "만원\n")
        
    def conference_day(self):
        print(self._name, "씨의", "이번 주 팀장회의는", self._manager_conference_day, "에 있습니다.\n")
        
class Staff(Employee):
    staff_num = 0
    
    def __init__(self, name, age, gender, hiredate, teamname, phone_num, salary, deadline):
        Employee.__init__(self, name, age, gender, hiredate, teamname, phone_num)
        self._salary = salary
        self._deadline = deadline
        Staff.staff_num += 1
        print(self._name, "님이", self._teamname, "팀에 입사하였습니다.", "\n총", Staff.staff_num, "명의 직원이 근무하게 되었습니다.\n")
        
    def increase_salary(self, val):
        self._salary += val
        print(val, "만원", "월급 인상!^^\n")
        
    def decrease_salary(self, val):
        self._salary -= val
        print(val, "만원", "월급 삭감!ㅜ\n")
        
    def get_salary(self):
        return self._salary
    def set_salary(self, val):
        self._salary = val
    salary = property(get_salary, set_salary)
    
    def aboutMe(self):
        Employee.aboutMe(self)
        print("월급 :", self._salary, "만원\n")
        
    def report_deadline(self):
        print(self._name, "씨의", "보고서 마감 기한은", self._deadline, "까지 입니다.\n")

class Hourly(Employee):
    hourly_num = 0
    
    def __init__(self, name, age, gender, hiredate, teamname, phone_num, hourly_wage, offday):
        Employee.__init__(self, name, age, gender, hiredate, teamname, phone_num)
        self._hourly_wage = hourly_wage
        self._offday = offday
        Hourly.hourly_num += 1
        print(self._name, "님이", self._teamname, "팀에 입사하였습니다.", "\n총", Hourly.hourly_num, "명의 알바가 근무하게 되었습니다.\n")
        
    def increase_hourly_wage(self, val):
        self._hourly_wage += val
        print(val, "원", "시급 인상!^^\n")
        
    def decrease_hourly_wage(self, val):
        self._hourly_wage -= val
        print(val, "원", "시급 삭감!ㅜ\n")
        
    def get_hourly_wage(self):
        return self._hourly_wage
    def set_hourly_wage(self, val):
        self._hourly_wage = val
    hourly_wage = property(get_hourly_wage, set_hourly_wage)
    
    def aboutMe(self):
        Employee.aboutMe(self)
        print("시급 :", self._hourly_wage, "원\n")
        
    def offday(self):
        print(self._name, "씨는", self._offday, "이 휴무일입니다.\n")
 
```

```python
%run Employee_class.py

manager1 = Manager("Cathy", 25, "여", "2019.05.03", "마케팅", "010-1234-5678", 300, '수요일')
manager1.aboutMe()

manager1.startWork()                               # decorator를 이용하여 현재시간이 같이 출력되도록 하였다.
print(manager1.name, "의 나이는", manager1.age)    # @property를 이용하여 name과 age을 attribute처럼 사용하였다. 
manager1.conference_day()
manager1.increase_salary(20)
manager1.endWork()
print(manager1.name, "의 현재 월급은", manager1.salary, "만원")   #변동된 salary 확인

print('---------------------------------------------------------------------')

manager2 = Manager("Mike", 30, "남", "2019.02.03", "개발", "010-4321-5678", 280, '목요일')
manager2.aboutMe()

manager2.startWork()                              # decorator를 이용하여 현재시간이 같이 출력되도록 하였다.
print(manager2.name, "의 나이는", manager2.age)   # @property를 이용하여 name과 age을 attribute처럼 사용하였다. 
manager2.conference_day()
manager2.decrease_salary(30)
manager2.endWork()
print(manager2.name, "의 현재 월급은", manager2.salary, "만원")  #변동된 salary 확인
```
