# 실습2

```python
import matplotlib.pylab as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
plt.plot(x, y)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ['$-\pi$', '$-\pi$/2', '0', '$\pi$/2', '$\pi$'])
plt.yticks([-1, 0, 1])
plt.show()
```

# 실습3

```python
import matplotlib.pylab as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 256)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ['$-\pi$', '$-\pi$/2', '0', '$\pi$/2', '$\pi$'])
plt.yticks([-1, 0, 1])
plt.show()
```
