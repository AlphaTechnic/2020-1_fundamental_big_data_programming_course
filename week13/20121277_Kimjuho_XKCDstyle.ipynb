```python
import matplotlib.pylab as plt
import numpy as np

plt.xkcd()
plt.title('XKCD style plot!!!')   # 타이틀 생성
plt.yticks(np.arange(-1,1.25,0.25))   # yticks를 default보다 더 세밀하게

x = np.linspace(-np.pi, np.pi, 256)       # x와 y 생성
y = np.cos(x)

plt.plot(x,y)
plt.plot(2*np.pi/3, -0.5, 'ro')            # 빨간 점 찍기
plt.annotate('0.5 Here!', xy = (2*np.pi/3, -0.5), xytext = (0.55, 0.05), textcoords = 'axes fraction', fontsize = 16,\
             arrowprops =  dict(arrowstyle = '->'))          # annotate

plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np

plt.xkcd()

fig = plt.figure()
ax = plt.subplot(111)

ax.spines['right'].set_visible(False)        #right 과 top의 축 제거
ax.spines['top'].set_visible(False)

plt.suptitle("\"Stove Ownership\" from xkcd by Randall Monroe", y =0.03 , size = 14)     # title을 아래에 위치시키고, size를 조정

plt.xticks([])                    # xticks, yticks 제거
plt.yticks([])
ax.set_ylim([0, 1.3])             # y=1인 상수함수를 그려줄 것이기 때문에, 예쁘게 그려주기 위해 그래프가 
                                  # 그려지는 y 범위를 0~1까지가 아닌 0~1.3까지로 조정

plt.xlabel('time')                    
plt.ylabel('my overall health')

x1 = range(0,7)                   # x1 = [0, 1, 2, 3, 4, 5, 6]
x2 = range(6,10)                   # x2 = [6, 7, 8, 9]
y1 = [1 for x in range(7)]         # y1 = [1, 1, 1, 1, 1, 1, 1]
y2 = [1, 0.66, 0.33, 0]             

plt.plot(x1, y1, 'b')              # x가 6일 때까지는 y=1이라는 상수함수를 그린다.
plt.plot(x2, y2, 'b')              # x가 6에서 9 까지 진행할 때, y는 1에서 0으로 감소하는 직선으로 그려짐.
plt.annotate("THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED", xy=(6,1), xytext = (1,0.67), \
             arrowprops = dict(arrowstyle = '->'))  # annotate

plt.show()
```
